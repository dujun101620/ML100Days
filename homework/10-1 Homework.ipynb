### 1. [簡答題] 請問下列四種不同的 DataFrame 選取結果有什麼差異？

```
df.loc[ '2013-01-01', 'A'] 
df.loc[ '2013-01-01', ['A', 'B'] ] 
df.loc[ '2013-01-01':'2013-01-02', 'A' ] 
df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] 
```





````python
#df.loc[ '2013-01-01', 'A'] :選取2013-01-01列A欄的資料
#df.loc[ '2013-01-01', ['A', 'B'] ] :選取2013-01-01列A欄與B欄的資料
#df.loc[ '2013-01-01':'2013-01-02', 'A' ]:選取2013-01-01列到2013-01-02列(不含2013-01-02列)A欄的資料
#df.loc[ '2013-01-01':'2013-01-05', 'A':'C'] :選取2013-01-01列到2013-01-05列(不含2013-01-05列)A欄到C欄(不含C欄)的資料
````

### 2. 請根據提供的資料，選擇出下列的要求：

```
# - select the first 3 rows.
# - select the odd rows. (index = 1, 3, 5)
# - select the last 2 columns.
# - select the even columns. (index = 0, 2, 4)
```


````python
# 記得先 Import 正確的套件
import numpy as np
import pandas as pd
````

````python
df = pd.DataFrame(np.random.randn(6,3),columns=['col1','col2','col3'])
df
````

````python
# - select the first 3 rows.
df1=df[0:3]
df1
````

````python
# - select the odd rows. (index = 1, 3, 5)
df2=df.loc[[1,3,5]]
df2
````

````python
# - select the last 2 columns.
df3=df.iloc[:,[-2,-1]]
df3
````

````python
# - select the even columns. (index = 0, 2, 4)
df4=df.iloc[[0,2,4],[0,2]]
df4
````

### 3. 請根據提供的資料，選擇出下列的要求：

```
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100
```

````python
df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
df
````

````python
# - 1. filtered by first column > 20?
df1=df[df[0]>20]
df1
````

````python
# - 2. filtered by first column + second column > 50
df2=df[df[0]+df[1]>50]
df2
````

````python
# - 3. filtered by first column < 30 or second column > 30
df3=df[(df[0]<30)|(df[1]>30)]
df3
````

````python
# - 4. filtered by total sum of row > 100
df4=df[df.sum(axis=1)>100]
df4
````
